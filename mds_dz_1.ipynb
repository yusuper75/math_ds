Найти косинус угла между векторами x1 и x2, где x1 и x2 - это два дома из 2ух атрибутов, полученных на этапе 2.

```python
x1 = np.array(x3)
x2 = np.array(x4)
```

```python
from numpy.linalg import norm
cos_angle = np.dot(x1, x2) / norm(x1) / norm(x2)
print ('Косинус угла между x1 и x2:', cos_angle)
print ('Сам угол:', np.arccos(cos_angle))
```
